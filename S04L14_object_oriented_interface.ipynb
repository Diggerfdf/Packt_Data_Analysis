## 3.2 Object Oriented Interface

What you do when plotting with the object oriented interphase, is to create the objects and then call methods for every object to make changes specifically to that object. The plt.subplots function creates a figure object and a Subplot(Axes) objesct at the same time, then we can create a plot by calling methods on those objects. 

```python
x = np.linspace(-4, 4)
y1, y2 = x ** 2, x ** 3
```

```python
# Object Oriented Interface
fig, ax = plt.subplots()
ax.plot(x, y1, 'red')
ax.plot(x, y2, 'blue')
ax.set_title('Quadratic and Cubic Powers')
ax.set_xlabel('x values')
ax.set_ylabel('y values');
```

Interestingly, just about all methods of an *Axes* object exist as function in the *pyplot* module (and vice-versa). For example, when calling plt.xlabel("x label"), *pyplot* calls ax.set_xlabel("x label") on wichever *Axes* is "current". Here is an equivalent version of the above example using just pyplot. 

```python
# pyplot
plt.plot(x, y1, 'red', x, y2, 'blue')
plt.title('Quadratic and Cubic Powers')
plt.xlabel('x values')
plt.ylabel('y values');
```

The usefulness of the OOi is clearer when working with multiple subplots in one figure. We've mentioned before that a figure can have more than on Axes on it. If you want your axes to be on a regular grid system, then it's easiest to use plt.subplots(...) to create a figure and add the axes to it automatically

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
```

plt.subplots(...)created a new figure and added 4 subplots to it. The axes object that was returned is a 2D numpy object array. Each item in the array is one of the subplots. They're laid out as you see them on the figure. Therefore, when we want to work with one of these axes, we can index the axes array and use that items methods. For example: 

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0,0].set_title('Upper Left')
axes[0,1].set_title('Upper Right')
axes[1,0].set_title('Lower Left')
axes[1,1].set_title('Lower Right')

# To iterate over all items in a multidimensional numpy array, use the 'flat' attribute
for ax in axes.flat:
    #remove all xticks and yticks...
    ax.set(xticks = [], yticks = [])
    
fig.tight_layout();
```

## 3.2.1 Example: Creating a figure showing the shape of the power functions.

```python
x = np.linspace(start=-5, stop=5, num=150)
```

```python
# All functions in one axes
fig, ax = plt.subplots(figsize = (7,4))
ax.plot(x, x, label='Linear')
ax.plot(x, x**2, label='Quadratic')
ax.plot(x, x**3, label='Cubic')
ax.plot(x, x**4, label='4th Power')
ax.legend();
```

```python
# 4 Axes/Subplots: One function in one axes
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (7, 4.5))
axes[0,0].set_title('Linear')
axes[0,0].plot(x, x)
axes[0,1].set_title('Quadratic')
axes[0,1].plot(x, x ** 2)
axes[1,0].set_title('Cubic')
axes[1,0].plot(x, x ** 3)
axes[1,1].set_title('4th Power')
axes[1,1].plot(x, x ** 4)
fig.tight_layout();
```

What if we want to take a look at the first 10 powers of x?

```python
# A more elegant approach
fig, axes = plt.subplots(nrows=2, ncols=5, figsize = (14, 4.5))
for i, ax in enumerate(axes.flatten()):
    ax.set_title("x to the {:d}".format(i+1))
    ax.plot(x, x ** (i+1))
    
fig.tight_layout();
```
